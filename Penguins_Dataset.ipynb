```python
cleaned_penguin_df = penguins.dropna()
cleaned_penguin_df
```

```python
cleaned_penguin_df.isna().sum()
```

```python
sns.scatterplot(data = cleaned_penguin_df, x="flipper_length_mm", y="body_mass_g",hue='island')
plt.title("Flipper Length vs Body Mass")
plt.show()
#  generates a scatter plot to visualize the relationship between the flipper length and body mass of the penguins.
#  representation of how a penguin's flipper length relates to its body mass
```

```python
sns.pairplot(data=cleaned_penguin_df,hue='species')
plt.suptitle("Pair Plot of Penguin Features", y=1.02)
plt.show()
# relationships between all the numerical columns in the dataset.
```

```python
sns.histplot(data=cleaned_penguin_df, x="body_mass_g", hue="species", kde=True)
plt.title("Body Mass Distribution")
plt.show()
# show how the body mass of penguins is distributed, with the distributions for each species overlaid on the same plot
#  compare the typical body mass and the range of body masses for Adelie, Chinstrap, and Gentoo penguins.
```

```python
sns.boxplot(data=cleaned_penguin_df, x="species", y="body_mass_g", hue="sex")
plt.title("Boxplot")
plt.show()
# compare the typical body mass and the range of body masses for Adelie, Chinstrap, and Gentoo penguins.
```

```python
sns.violinplot(data=cleaned_penguin_df, x="species", y="flipper_length_mm", hue="sex", split=True)
plt.title("Violin Plot")
plt.show()
Each "violin" shows:

# The wider sections of the violin: These represent a higher probability density for the data at that value. In other words, where the violin is wider, there are more data points.
# The narrower sections: These represent a lower probability density.
# The plot inside the violin: This is a miniature box plot, showing the median, interquartile range, and whiskers, just like in a regular boxplot.
```

```python
sns.barplot(data=cleaned_penguin_df, x="island", y="bill_length_mm")
plt.title("Average Bill Length per Island")
plt.show()
# The height of each bar represents the average bill length of all the penguins in the dataset that were found on that specific island. The small black lines at the top of the bars are error bars, which give an indication of the uncertainty in the calculated mean.
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=cleaned_penguin_df, x="species", hue="sex")
plt.title("Countplot")
plt.show()
#  The height of each bar represents the total number of penguins that fall into that specific category combination
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=cleaned_penguin_df, x="species", hue="island")
plt.title("Number of Penguins by Species and Island")
plt.show()
# The height of each bar shows the number of penguins of a particular species that were observed on a specific island.
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(data=cleaned_penguin_df, x="species", y="body_mass_g", hue="sex")
plt.title("Body Mass by Species and Sex")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.swarmplot(data=cleaned_penguin_df, x="species", y="bill_length_mm", hue="sex")
plt.title("Bill Length by Species and Sex")
plt.show()
```
